```python
import pandas as pd
import numpy as np
df=pd.read_csv('E:/ml/energydata_complete.csv')
```

```python
df.shape
```

```python
df.dropna()
```

```python
df = df.replace([np.nan, -np.inf], 0)
```

```python
df.head()
```

```python
x_mean=df.mean() #finding mean,std,variance
x_std=df.std()
z=(df-x_mean)/x_std
```

```python
c=z.cov()     #covariance

import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(c)
plt.show()
```

```python
eigenvalues, eigenvectors = np.linalg.eig(c)
print('Eigen values:\n', eigenvalues)
print('Eigen values Shape:', eigenvalues.shape)
print('Eigen Vector Shape:', eigenvectors.shape)
```

```python
idx=eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors =eigenvectors[:,idx]
```

```python
explained_var= np.cumsum(eigenvalues)/np.sum(eigenvalues)
explained_var
```

```python
n_components = np.argmax(explained_var >=0.50)+1
n_components
```

```python
u = eigenvectors[:,:n_components]
pca_component = pd.DataFrame(u,index=["date","Appliances","lights","T1","RH_1","T2","RH_2","T3","RH_3","T4","RH_4","T5","RH_5","T6","RH_6","T7","RH_7","T8","RH_8","T9","RH_9","T_out","Press_mm_hg","RH_out","Windspeed","Visibility","Tdewpoint","rv1"
],columns =['Pc1','Pc2'])
```

```python
plt.figure(figsize=(5,7))
sns.heatmap(pca_component)
plt.title('PCA Component')
plt.show()
```
