```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("E:/125015049/cervcancer.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.title('Cervical cancer Behaviour risk')
sns.distplot(df['behavior_sexualRisk'])
plt.show()
```

```python
plt.scatter(df['behavior_sexualRisk'],df['behavior_eating'])
plt.title('Sexual risk vs eating')
plt.xlabel('sexual risk')
plt.ylabel('eating')
plt.box(False)
plt.show()
```

```python
X = df.iloc[:, :1]  # independent
y = df.iloc[:, 1:]  # dependent
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_test = regressor.predict(X_test)     # predicted value of y_test
y_pred_train = regressor.predict(X_train)   # predicted value of y_train
```

```python
# Prediction on training set
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```
